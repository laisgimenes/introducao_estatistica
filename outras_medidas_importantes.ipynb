```python
import pandas as pd
import numpy as np
import scipy.stats as stats
from scipy.stats import variation
from statistics import variance
from scipy.stats import skew
from scipy.stats import kurtosis
```

```python
dados = pd.read_excel('tabela_2_1.xlsx')
dados
```

```python
dados.columns
```

```python
# Renomeando as colunas

novos_nomes_colunas = {'Estado Civil':'estado_civil','Grau de Instrução':'escolaridade','N de Filhos':'qtd_filhos',
                      'Salario (x Sal Min)':'salario','Anos':'idade_anos','Meses':'idade_meses',
                       'Região de Procedência':'regiao'}

dados.rename(columns=novos_nomes_colunas, inplace = True)
dados
```

```python
dados.columns
```

## Medidas de Posição Relativa

- Percentis, Quartis, Decis e Escore Z

```python
# Escore Z

stats.zscore(dados['salario'])
```

```python
# Percentis ou Centis

dados['salario'].quantile([i/100 for i in range(1,99)])
```

```python
# Decis

dados['salario'].quantile([i/10 for i in range(1,10)])
```

```python
# Quartis

dados['salario'].quantile([0.25,0.50,0.75])
```

## Amplitude

```python
amplitude = max(dados['idade_anos']) - min(dados['idade_anos'])
print(f'amplitude da variável idade em anos: {amplitude} ')
```

## Intervalo Interquartil

```python
# Interquartile range (IQR)
IQR = stats.iqr(dados['salario'], interpolation = 'midpoint')
print(f'Intervalo Interquartil da variável salário: {IQR}')


# ou

Q3 = dados['salario'].quantile([0.75])
Q1 = dados['salario'].quantile([0.25])

print(f'Intervalo Interquartil da variável salário: {float(Q3.values) - float(Q1.values)}')
```

## Coeficiente de Variação

```python
print(variation(dados['salario'])*100)

print(np.std(dados['salario']) / np.mean(dados['salario'])*100)
```

## Medidas de Assimetria e Curtose

```python
# Assimetria

print(skew(dados['salario']))

#Curtose

print(kurtosis(dados['salario']))
```
